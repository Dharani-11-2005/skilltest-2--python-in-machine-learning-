```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create sample data
dates = pd.date_range(start='2025-01-01', periods=365)
temps = np.random.normal(25, 5, size=365)

df = pd.DataFrame({'Date': dates, 'Temperature': temps})
df.set_index('Date', inplace=True)

# Rolling mean
df['Rolling Mean'] = df['Temperature'].rolling(window=7).mean()

# Plot
df.plot()
plt.title('Time Series Temperature')
plt.show()
```
