```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(0)
days = 365
price = 100
prices = [price]
for i in range(days):
    price += np.random.normal(0, 2)
    prices.append(price)
df = pd.DataFrame({'Price': prices})
# Moving average
df['MA_10'] = df['Price'].rolling(10).mean()
# Plot
df.plot()
plt.title('Stock Price Simulation')
plt.show()
```
