```python
import pandas as pd

df = pd.read_excel('student_scores.csv.xlsx')

# Summary statistics
print(df.describe())

# Average score for each student
df['Average'] = df[['Maths', 'Physics', 'Chemistry']].mean(axis=1)
print(df[['Name', 'Average']])

# Students scoring below 60 in more than 2 subjects
low_scores = (df[['Maths','Physics','Chemistry']] < 60).sum(axis=1)
result = df[low_scores > 2]

print(result)
```
