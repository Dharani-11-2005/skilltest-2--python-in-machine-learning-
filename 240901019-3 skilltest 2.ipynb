```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_excel('flight_data.csv.xlsx')

# Summary statistics of delay
print(df['Departure_Delay'].describe())

# Average delay per airline
avg_delay = df.groupby('Airline')['Departure_Delay'].mean()
print(avg_delay)

# Most common routes
print("Most common departure:", df['Origin'].mode()[0])
print("Most common arrival:", df['Destination'].mode()[0])

# Box plot
df['Departure_Delay'].plot(kind='box')
plt.title('Departure_Delay Distribution')
plt.show()
```
